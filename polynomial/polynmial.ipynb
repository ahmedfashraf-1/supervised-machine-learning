```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradient_descent(X, y, iterations, alpha):
    m = len(y)
    n = X.shape[1]
    w = np.zeros(n)
    b = 0.0
    for i in range(iterations):
        prediction = np.dot(X, w) + b
        error = prediction - y
        dw = (1 / m) * np.dot(X.T, error)
        db = (1 / m) * np.sum(error)
        w -= alpha * dw
        b -= alpha * db
        if i % 100 == 0:
            cost = (1 / (2 * m)) * np.sum(error ** 2)
            print(f"Iteration {i}, Cost: {cost:.4f}")
    return w, b
```

```python
x=np.arange(0,20,1)
y=1+x**2
X=x.reshape(-1,1)#convert to 2d matrix
model_w,model_b=gradient_descent(X,y,iterations=1000,alpha=0.01)
plt.scatter(x,y,marker='x',c='r',label='actual values')
plt.title("no feature engineering")
plt.plot(x,np.dot(X, model_w) + model_b,label='predicted values')
plt.legend()
plt.show()
```

```python
def standardize(X):
    mean =X.mean(axis=0)
    std =X.std(axis=0)
    return (X - mean) / std
```

```python
x = np.arange(0,20,1)
y = x**2
X = np.c_[x, x**2, x**3]
X = standardize(X)
model_w, model_b = gradient_descent(X, y, iterations=100000, alpha=1e-1)
plt.scatter(x, y, marker='x', c='r', label="Actual Value")
plt.title("Normalized x x**2, x**3 feature")
plt.plot(x,X@model_w + model_b, label="Predicted Value")
plt.legend()
plt.show()
```
